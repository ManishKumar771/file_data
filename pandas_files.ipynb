```python

import pandas as pd
import io
import requests

print("-----csv-----")

url = "https://raw.githubusercontent.com/ManishKumar771/file_data/main/file.csv"

df = pd.read_csv(url)
print(df)


print("\n-----json-----")


url01 = "https://raw.githubusercontent.com/ManishKumar771/file_data/main/data01.json"

ds = pd.read_json(url01)
print(ds)





print("\n-----excel-----")

url02 = "https://raw.githubusercontent.com/ManishKumar771/file_data/main/file01.xlsx"


response = requests.get(url02)
dx = pd.read_excel(io.BytesIO(response.content))
print(dx)


```
